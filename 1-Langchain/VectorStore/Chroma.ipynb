# Chroma

Chroma is a AI-native open-source vector database focused on developer productivity and happiness. Chroma is licensed under  Apache 2.0

```python
# builder a sample vectordb

from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader,PyPDFLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter,RecursiveCharacterTextSplitter
```

```python
loader = PyPDFLoader("Abhishek Jaiswar - Web Developer.pdf")
data = loader.load()
data
```

```python
# Split
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
split = text_splitter.split_documents(data)
```

```python
embeddings =  OllamaEmbeddings(
    model="qwen3-embedding:0.6b"
)
vectordb = Chroma.from_documents(documents=split, embedding=embeddings)
vectordb
```

```python
## query it

query = "Projects ?"
docs = vectordb.similarity_search(query)
docs[0].page_content
```

```python
## Saving to the disk

vectordb = Chroma.from_documents(documents=split, embedding=embeddings,persist_directory="./chroma_db")


```

```python
db = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)

docs = db.similarity_search(query)
print(docs[0].page_content)
```

```python
## Retriever option

retriver = vectordb.as_retriever()

retriver.invoke(query)[0].page_content
```
