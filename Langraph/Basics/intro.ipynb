## Langraph Overview

LangGraph, created by Langchain, is an open source AI agent Framework designed to build,deploy and manage complex generative AI agent workflows. It provides a set of tools  and libraries that enable users to create, run and optimize large language models (LLMs) in a scalable and efficient manner. At its core, LangGraph uses the power of graph-based architecutes to model and manage the intricate relationships between variouse components of an AI agent workflow

- It is an orchestration framework for building intelligent,stateful and multi-step LLM workflows
- It enables adanced features like parallelism , loops , branching , memroy and resumablity - making it idal for agentic and production grade AI application
- It models your logic as a graph of nodes (tasks) and edeges(routing) insted of a linear chain

# LLM Workflows

- LLM workflows are a step by step process using which we can build complex LLM applications
- Each step in a workflow performs a distinct task - such as prompting, reasoning, tool calling, memory access, or decision-making
- workflows can be linear,parallel, branched or looped allowing for complex behaviours like retries, multi-agent communcation or tool-augmented reasoning

# Reducers

Reducers in Langraph define how updates from nodes are applied to the shared state.

Each key in the state can have its own reducer, which determines whether new data replaces, merges or adds to the existing value.

# LangGraph Execution Model - Google Pragel

1. Graph Definition
 - The state schema
 - Nodes (function that perform tasks)
 - Edges (which node connects to which)
  
2. Compilation
   you call .compile() on the StateGraph
   This checks the graph structure and prepares it for execution

3. Invocation
   You run the graph with .invoke(initial_state).
   LangGraph sends the inital state as a message to entry node

4. Super - steps Begin
   Execution proceeds in rounds

5. Message Passing & Node Activation
   The messages are passed to downstream nodes via edges
   Nodes that receive messages become active for the next round

6. Halting Condition
    Execution stops when:
    - No nodes are active and
    - NO Messages are in transit
