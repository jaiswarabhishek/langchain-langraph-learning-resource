```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["HF_TOKEN"] = os.getenv("HF_TOKEN")

## Langsmith Tracking
os.environ['LANGCHAIN_API_KEY'] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
```

```python
from langchain_openai  import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o")

print(llm)
```

```python
from langchain_community.llms import Ollama

# 5. LLM + prompt for generation
llm = Ollama(model="gemma3:1b")
```

```python
## input and get response from llm

result = llm.invoke("what is gen ai")
print(result)
```

```python
## Chat prompt template - tell llm how to behave , how to response etc.
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert AI Engineer. Provide me answers based on the question"),
        ("user","{input}")
    ]
)

print(prompt)
```

```python
chain = prompt | llm

response = chain.invoke({"input":"can you tell me about AI ?"})
print(response)
```

```python
# strouput Parser

from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
chain = prompt | llm | output_parser


response = chain.invoke({"input":"what day today?"})
print(response)
```
